```python
# Age demographic - we don't have this in our instruction! i did it because it shows on the starter ipynb
# bin to hold the age value 
bins = [0,10,15,20,25,30,35,40, 45]
age_range = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# purchase count by age group
data_file["Age Range"] = pd.cut(data_file["Age"], bins, labels=age_range)
data_file.head()

# age group percentage 
age_percentage_df = round(data_file["Age Range"].value_counts()/number_purchase,2)*100
age_percentage_df

# Total age count
age_count_df = data_file.groupby("Age Range")["SN"].nunique()
age_count_df.count()

# Summary table 
age_demographic_df = pd.DataFrame({"Total Count":age_count_df,
                               "Percentage of Players": age_percentage_df})
age_demographic_df

# Format table
age_demographic_df.style.format({"Percentage of Player":"{:,.2f}%"})
```

```python
# Purchasing Analysis (Age)
# Age count
age_purchase_df = data_file.groupby("Age Range")["Item Name"].count()
age_purchase_df.head()

# Average purchase price by age
Average_purchase_price_df = round(data_file.groupby("Age Range")["Price"].mean(),2)
Average_purchase_price_df

# Total purchase value
total_purchase_value_df = round(data_file.groupby("Age Range")["Price"].sum(),2)
total_purchase_value_df

# Average Purchase Total per Person by Age Group
per_person_df = round(total_purchase_value_df/age_count_df,2)
per_person_df

# summary table 
age_range_table_df = pd.DataFrame({"Purchase Count":age_purchase_df,
                            "Average Purchase Price":Average_purchase_price_df,
                            "Total Purchase Value": total_purchase_value_df,
                            "Avg Total Purchase per Person": per_person_df})
age_range_table_df

# format the table
age_range_table_df.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

```python
# Identify the the top 5 spenders in the game by total purchase value, then list (in a table)

spenders_count_df = data_file.groupby("SN")["Purchase ID"].count()
spenders_average_price_df = round(data_file.groupby("SN")["Price"].mean(),2)
spenders_total_df = data_file.groupby("SN")["Price"].sum()

# Summary table 
spender_table_df = pd.DataFrame({"Purchase Count":spenders_count_df,
                                   "Average Purchase Price":spenders_average_price_df,
                                   "Total Purchase Value":spenders_total_df})
spender_table_df.head()

# Table formatting
spender_table_df.style.format({"Average Purchase Price":"${:,.2f}", 
                              "Total Purchase Value":"${:,.2f}"})

# Determining the top 5 spenders
top_five_spenders = spender_table_df.sort_values("Total Purchase Value", ascending=False)
top_five_spenders.head()
```

```python
## Most Popular Items
# Identify the 5 most popular items by purchase count, then list (in a table)

items_purchase_count_df = data_file.groupby(["Item ID", "Item Name"])["Price"].count()
items_average_price_df = round(data_file.groupby(["Item ID", "Item Name"])["Price"].mean(),2)
items_value_total_df = data_file.groupby(["Item ID", "Item Name"])["Price"].sum()

# Summary Table

items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count_df,
                                   "Item Price":items_average_price_df,
                                   "Total Purchase Value":items_value_total_df,})

items_purchased_df.head()

# Table formatting
items_purchased_df.style.format({"Item Price":"${:,.2f}", 
                              "Total Purchase Value":"${:,.2f}"})

# Determining the five most popular items.

most_popular_items_df = items_purchased_df.sort_values("Purchase Count", ascending=False)
most_popular_items_df.head()
```
